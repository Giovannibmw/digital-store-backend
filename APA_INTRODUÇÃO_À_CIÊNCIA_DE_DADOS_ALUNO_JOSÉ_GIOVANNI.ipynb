```python
# 4º Exercício:  Criar uma matriz com 50 elementos e 03 dimensões.
import numpy as np

# Criando um array com 50 elementos e 3 dimensões
v = np.array(range(50)).reshape(2, 5, 5)

print('Shape = ', v.shape)
print('Número de dimensões = ', v.ndim)
print('Número de elementos = ', v.size)
print('Tensor v = \n', v)
```

```python
# 5º Exercício: Criar um gráfico pizza com 03 itens destacando um item.
import matplotlib.pyplot as plt

# Definir as variáveis
fatias = [8, 12, 10]
atividades = ['Lazer', 'Estudo', 'Viagem']
colunas = ['r', 'm', 'y']  # cores: red, magenta, yellow

# Criar o gráfico de pizza
plt.pie(
    fatias,                      # valores de cada fatia
    labels=atividades,           # nomes das categorias
    colors=colunas,              # cores das fatias
    startangle=90,               # começa o gráfico aos 90º
    shadow=True,                 # adiciona sombra
    explode=(0, 0.1, 0)          # destaca a 2ª fatia (Estudo)
)

# Taí o gráfico
plt.show()
```

```python
# 6º Exercício: Criar um gráfico tipo linha.
import matplotlib.pyplot as plt

# Definir as variáveis
x = [1, 3, 5]
y = [1, 2, 5]

# Criar o gráfico de linha
plt.plot(x, y, label='Uma legenda')  # adicionando a legenda já na criação

# Títulos e rótulos dos eixos
plt.title('Exemplo utilizando Plot')   # título do gráfico
plt.xlabel('Variável 1')               # nome do eixo X
plt.ylabel('Variável 2')               # nome do eixo Y

# Exibir a legenda
plt.legend()

# Olha aí o gráfico
plt.show()
```

```python
# 7º Exercício: Criar um gráfico com pontos em formato de estrela.
import matplotlib.pyplot as plt

# Definir variáveis
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 5, 6, 8, 4, 8]

# Criar o gráfico de dispersão com formato de estrela
plt.scatter(
    x, y,
    label='Pontos',
    color='b',    # 'b' = blue
    marker='*',   # formato de estrela
    s=100         # tamanho dos marcadores
)

# Exibir legenda e gráfico de estrelinhas :)
plt.legend()
plt.show()
```
